```python
#dropping unnecessary columns
data=data.drop(["Engine Fuel Type","Market Category","Vehicle Style","Popularity","Number of Doors","Vehicle Size"],axis=1)
print(data.head(5))
print(data.columns)
```

```python
#rename columns
data=data.rename(columns={"Engine HP":"HP","Engine Cylinders":"Cylinders","Driven_Wheels":"Wheels"})
print(data.head(5))
print(data.columns)
```

```python
#rows containing duplicate data and it's shape
duplicate=data[data.duplicated()]
print(duplicate)
print(duplicate.head(20))
print(duplicate.shape)
```

```python
#count the rows before removing the data
print(data.count())
```

```python
#drop the duplicates and count the number of rows
data=data.drop_duplicates()
print(data.head())
print(data.count())
```

```python
#find the null values
print(data.isnull().sum())
```

```python
#drop the missing values and count the number of rows
data=data.dropna()
print(data.count())
```

```python
#printing the null values after dropping the missing values
print(data.isnull().sum())
```

```python
#finding ouliers
print(sns.boxplot(x=data["MSRP"]))
print(plt.show())
print(sns.boxplot(x=data["HP"]))
print(plt.show())
```

```python
#histogram for number of cars per brand
print(data.Make.value_counts().nlargest(40).plot(kind="bar",figsize=(10,5)))
plt.title("Number of cars of different brands")
plt.xlabel("Brands")
plt.ylabel("Number of cars")
```

```python
#finding the relationships between the variables(columns)
plt.figure(figsize=(20,10))
c=data.corr()
sns.heatmap(c,cmap="BrBG",annot=True)
print(c)
```
